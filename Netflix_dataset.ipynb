### Que 1: For 'Zodiac' , what is the show id and who is the director of the show?

### method 1 : Using isin()

```python
data[data['title'].isin(['Zodiac'])]
```

###  method 2: Using str.contains()

str.contains() is used to show all records of a particular string in the column

```python
data[data['title'].str.contains('Zodiac')]
```

### Findings : 

show_id for movie 'Zodiac' :  **s8803**


director of movie 'Zodiac' :  **David Fincher**

### Que 2: In which year the highest number of the TV shows and movies were were added to netflix ? Show with Bar graph

```python
data.dtypes
```

```python
data['date_added'] = pd.to_datetime(data['date_added'])
```

```python
data.dtypes
```

```python
data['date_added'].dt.year.value_counts()
```

### Plotting the Bar graph

```python
data['date_added'].dt.year.value_counts().plot(kind = 'bar')
```

###  Findings : 

Maximum number of TV shows and Movies were added to netflix in year 2019

### Que 3: How many Movies and TV shows are in the dataset ? Show with Bar graph

### method 1:   groupby()

```python
data.groupby('type').type.count()          #count of only 'type' column
```

### method 2: countplot()

```python
sns.countplot(data['type'])
```

### Findings : 

Total number of Movies: 6131

Total number of TV Shows: 2676

###  Que 8: In how many Movies or shows Robert Downey Jr was cast?

For the purpose we need to use .str.contains() but as there are NA and NaN in the column so this method will throw error.

For resolving this we need to remove the NA and NaN from the column

```python
data_new = data.dropna()
```

```python
data_new[data_new['cast'].str.contains('Robert Downey Jr')]
```

### Findings : 
Netflix has 7 movies in which Robert Downey Jr is in the cast.
